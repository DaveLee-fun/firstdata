```python
import pandas as pd
import plotly.express as px

# 데이터 불러오기
df = pd.read_csv('seoul_apartment.csv')

# 계약년월별 평균 매매가격 계산
monthly_prices = df.groupby('계약년월')['거래금액(만원)'].mean()
monthly_prices
```

```python
fig = px.line(monthly_prices, x=monthly_prices.index, y=monthly_prices.values, title='Monthly Average Apartment Prices in Seoul')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# 데이터 불러오기
df = pd.read_csv('seoul_apartment.csv')

# 계약년월별 평균 매매가격 계산
monthly_prices = df.groupby('계약년월')['거래금액(만원)'].mean()

# 인덱스를 문자열로 변환
monthly_prices.index = monthly_prices.index.astype(str)

# 그래프 생성
fig = px.line(monthly_prices, x=monthly_prices.index, y=monthly_prices.values, title='Monthly Average Apartment Prices in Seoul')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('seoul_apartment.csv')

transactions = df.groupby('계약년월')['NO'].count()

# 인덱스를 문자열로 변환
transactions.index = transactions.index.astype(str)

fig = px.bar(transactions, x=transactions.index, y=transactions.values, 
             title='서울 아파트 계약년월별 거래 횟수',
             labels={'x': '계약년월', 'y': '거래 횟수'})
fig.show()
```
